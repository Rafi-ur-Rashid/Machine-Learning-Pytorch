```python
import numpy as np
import pandas as pd
```

```python
data_path="/content/drive/MyDrive/Reserach_Notebooks/data/"
```

## Preprocessing IMdB data

```python
df = pd.read_csv(data_path+'imdb/IMDB Dataset.csv')
texts=df.review.tolist()

all_texts=''
for t in texts:
  all_texts= all_texts+' '+t

with open(data_path+'imdb/all_texts.txt', 'w') as file:
  file.write(all_texts)
```

## Preprocessing CleanJokes data

```python
df = pd.read_csv(data_path+'reddit_clean_jokes/reddit-cleanjokes.csv')
texts=df.Joke.tolist()

all_texts=''
for t in texts:
  all_texts= all_texts+' '+t
```

```python
with open(data_path+'reddit_clean_jokes/all_texts.txt', 'w') as file:
  file.write(all_texts)
```

## Preprocessing WikiText2 data

```python
with open(data_path+'wiki/texts.txt') as file:
  texts=[line.rstrip() for line in file]

all_texts=''
for t in texts:
  if "= =" not in t:
    all_texts= all_texts+t
```

```python
with open(data_path+'wiki/all_texts.txt', 'w') as file:
  file.write(all_texts)
```

## Preprocessing ptb data

```python
with open(data_path+'ptb/texts.txt') as file:
  texts=[line.rstrip() for line in file]

all_texts=''
for t in texts:
  all_texts= all_texts+t
```

```python
with open(data_path+'ptb/all_texts.txt', 'w') as file:
  file.write(all_texts)
```
